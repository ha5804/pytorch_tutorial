# __Graph__

- Graph의 개념은 1,2장에서 설명하였다. 좀 더 구체적인 내용을 살펴보자.

## __1.grad.fn__
- Computation graph : 그래프는 각 Tensor가 자기의 생성 이력(grad_fn)과 연결 정보를 가지고 있는 형태로 붙어있다.
    `x -> mulBackward -> y -> powbackward -> z`

    x : leaf_bode

    y : mulBackward 노드와 연결

    z : powBackward 노드와 연결

    따라서, y.grad_fn, z.grad_fn은 각각 이전의 노드로 거슬러 올라가는것이다.

    `즉 grad_fn은 backward 계산을 수행할 autograd 노드 객체이다.`

- grad_fn : grad_fn은 autograd 노드 객체라고 설명하였다.
    - 이 텐서가 어떤 연산을 통해 생성되었는지.
    - 이전 노드가 무엇인지.
    - backward 때 필요한 중간값이 뭔지

    이런 정보를 가지고, 내가 backward될 때 어떻게 미분할지 아는 객체이다.


   

```python
import torch
x = torch.tensor(2.0, requires_grad = True)
y = 3 * x
z = y ** 2
z.backward()

print(y.grad_fn)
print(z.grad_fn)
```

- (3)next_functions : backward node는 보통 이전 노드들로 가는 포인터를 갖는다.


```python
print(z.grad_fn.next_functions)
#z이전 연산은 y의 연산으로 y가 어떤 연산으로 이루어졌는지 표현
```
