# Tensor

Tensor is structure like numpy array. 

But main difference tensor - ndarray is GPU.

#### (1) lets make array with tensor

    torch.tensor



```python
import torch
import numpy as np

data = [[1,2], [3,4]]
x_data = torch.tensor(data)

print(x_data)
```

* it can change ndarray - tensor

```python
x_data = np.array(x_data)
print(x_data)
x_data = torch.tensor(x_data)
print(x_data)
```

```python
shape = (2, 3)
ones_tensor = torch.ones(shape)
rands_tensor = torch.rand(shape)
zeros_tensor = torch.zeros(shape)
print(ones_tensor)
print(rands_tensor)
print(zeros_tensor)
```

#### (2) Operation

```python
first_row = ones_tensor[0]
first_col =ones_tensor[:, 0]
last_col = ones_tensor[:, -1]
print(first_row)
print(first_col)
print(last_col)
```

```python
tensor_cat = torch.cat([ones_tensor,ones_tensor,ones_tensor],dim = 1)
print(tensor_cat)
```
